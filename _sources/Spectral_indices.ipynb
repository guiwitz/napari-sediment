## ROI

For large images it is often not feasible or desirable to compute the index for the full image. Also, one may wish to compute indices in a narrow band in order to average the index along the x axis and be able to have a plot of index vs. depth. For this one can set a ROI in the ROI tab. A width can be indicated and then double-clicking on the image will generate the roi.

````python
nbscreenshot(viewer)
````
